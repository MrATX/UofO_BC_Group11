```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
vbm_states = ["Oregon","Washington","Colorado","Utah","Hawaii"]
or_yrs = []
wa_yrs = []
co_ut_yrs = []
for i in range(1980,2017,2):
    or_yrs.append(i)
for i in range(1994,2017,2):
    wa_yrs.append(i)
for i in range(2010,2017,2):
    co_ut_yrs.append(i)
file = "voter_data.csv"
voter_data = pd.read_csv(file)
or_slice = voter_data.loc[(voter_data["Year"].isin(or_yrs)),:]
or_slice = or_slice.loc[or_slice["State"]=="Oregon",:]
or_slice.sort_values(by=['Year'], inplace=True)
wa_slice = voter_data.loc[voter_data["Year"].isin(wa_yrs),:]
wa_slice = wa_slice.loc[wa_slice["State"]=="Washington",:]
wa_slice.sort_values(by=['Year'], inplace=True)
co_slice = voter_data.loc[voter_data["Year"].isin(co_ut_yrs),:]
co_slice = co_slice.loc[co_slice["State"]=="Colorado",:]
co_slice.sort_values(by=['Year'], inplace=True)
ut_slice = voter_data.loc[voter_data["Year"].isin(co_ut_yrs),:]
ut_slice = ut_slice.loc[ut_slice["State"]=="Utah",:]
ut_slice.sort_values(by=['Year'], inplace=True)
avg_nonvbm = voter_data.loc[~voter_data["State"].isin(vbm_states)]
avg_nonvbm = avg_nonvbm.groupby("Year").mean()
years = voter_data.loc[:,"Year"]

yticks = np.arange(0.35,0.85,0.10)
a = 0
for i in yticks:
    i = round(i,2)
    yticks[a] = i
    a = a + 1
```

```python
#Oregon
fig1, ax1 = plt.subplots()
ax1.set_xticks(or_yrs)
ax1.set_xticklabels(or_yrs,rotation="vertical")
ax1.set_title("Oregon Voter Turnout")
ax1.set_ylabel('Voter Turnout % (EVP)')
ax1.set_xlabel('Year')
ax1.axvline(x=1998, ymin=0.0, ymax=1, color="gold")
ax1.plot(avg_nonvbm.index,avg_nonvbm["VEP Total Ballots Counted"],color="black")
ax1.plot(voter_data.loc[voter_data["State"]=="Oregon","Year"],voter_data.loc[voter_data["State"]=="Oregon","VEP Total Ballots Counted"],color="blue")
ax1.set_yticks(yticks)
ax1.set_yticklabels(yticks)
ax1.legend(["Year of VBM Intro","Non-VBM State AVG","Oregon"],loc="lower left")
plt.show()
```
